```python
def prep(name):
    with open(f"data/{name}.txt", 'r') as file:
        data = file.readlines()
        data_clean = []
        for e in data:
            data_clean.append(e.strip())
    return data_clean
```

```python
data = prep("palindromi")
```

```python
type(data[0])
```

```python
def is_palindrom(text: str)->bool:
    return text == text[::-1]
```

```python
padlindromes = []
for e in data:
    if is_palindrom(e) and len(e) > 10:
        padlindromes.append(e)
```

```python
padlindromes[0]
```

```python
lst_dict = []
for i in padlindromes:
    rijecnik = {}
    for c in i:
        rijecnik[c] = 0

    for c in i:
        rijecnik[c] += 1

    lst_dict.append(rijecnik)
```

```python
lst_dict
```

```python
good_dicts_lst = []
for i in lst_dict:
    more_than_2 = False
    for e in i.values():
        if e > 2: more_than_2 = True
    if more_than_2:
        continue
    good_dicts_lst.append(i)
```

```python
def is_palindrome(s):
    return s == s[::-1]

def has_unique_character_limit(s):
    from collections import Counter
    char_counts = Counter(s)
    return all(count <= 2 for count in char_counts.values())

def count_palindromic_strings(filename):
    count = 0
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.strip()
            if len(line) > 10 and is_palindrome(line) and has_unique_character_limit(line):
                count += 1
    return count

# Upotrijebi naziv datoteke "palindromi.txt"
filename = 'data/palindromi.txt'
result = count_palindromic_strings(filename)
print(f"Broj palindromskih stringova koji zadovoljavaju uvjete: {result}")
```

```python
def get_ascii(char):
    rez = ord(char) - 10
    if rez < ord('a'):
       rez = (ord('z')+ (rez - ord('a')) + 1)
    return chr(rez)
```

```python
get_ascii('z')
```

```python
def transform_str(line:str)->str:
    new_line = ""
    for l in line:
        new_line += get_ascii(l)
    return new_line
```

```python
new_data = []
for e in data:
    new_data.append(transform_str(e))
```

```python
new_data[0]
```

```python
samoglasnici = {'a', 'e', 'i', 'o', 'u'}
```

```python
result = []
for e in new_data:
    if set(e) & samoglasnici:
        result.append(e)
```

```python
def get_interval(string: str) -> tuple:
    lower = string[0]

    lower_i = lower
    higher_i = lower
    for i in range(len(string) - 1):
        lower_i += '0'
        higher_i += '9'
    
    return (lower_i, higher_i)
```

```python
get_interval("1618")
```

```python
intervali = []
for i in data:
    intervali.append(get_interval(i))
```

```python
intervali
```

```python
rijecnik = {}
for i in set(intervali):
    rijecnik[i] = 0
rijecnik
```
